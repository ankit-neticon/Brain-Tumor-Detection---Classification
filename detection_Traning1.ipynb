```python
import cv2
from matplotlib import pyplot as plt
from PIL import Image 
import numpy as np
img = cv2.imread('Tumor_test/meningioma_tumor/image(26).jpg',0)

print(img.shape)
height,width=img.shape
plt.imshow(img,cmap='gray')
print(img)
```

```python
imgFlatten = img.reshape(height*width,1)
print(imgFlatten.shape)
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=6)
model.fit(imgFlatten)
labels = model.labels_
print(labels.shape)
labels2D=labels.reshape(height,width)
```

```python
plt.imshow(labels2D)
plt.colorbar()
plt.show()
```

```python
print(img[200:210,200:210])
print('------------------------------------------------------------')
print(labels2D[200:210,200:210])
```
