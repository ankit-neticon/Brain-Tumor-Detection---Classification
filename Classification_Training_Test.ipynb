```python
import joblib
model = joblib.load('brain_tumor_kMeans_module.sav')
tumorLabel=4.0
```

```python
from cv2 import *
from matplotlib import pyplot as plt 
img= cv2.imread('Tumor_test/glioma_tumor/image(22).jpg',0)
img_original = cv2.imread('Tumor_test/glioma_tumor/image(22).jpg')
#img= cv2.imread(img_path,0)
height,width = img.shape
imgFlatten = img.reshape(height*width,1)
labels = model.predict(imgFlatten)
labels2D = labels.reshape(height, width)
mask = (labels2D==tumorLabel)
plt.imshow(labels2D)
```

```python
from matplotlib import pyplot as plt
import numpy as np
tumorExtracted = np.bitwise_and(mask,img)
```

```python
from cv2 import *
def detection():
    contours,hieararchy = cv2.findContours(tumorExtracted,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    print('no. of contours: ', len(contours))
    #cv2.drawContours(img_original,contours,-1,(0,255,255),2)
    for index,cnt in enumerate(contours):
        area=cv2.contourArea(cnt)
        #print(area,index)

        font = cv2.FONT_HERSHEY_SIMPLEX 
        fontScale = 1 
        color = (255, 0, 0) 
        thickness = 2

        if(area>500):
            cv2.drawContours(img_original,[cnt],-1,(0,255,255),2)
            x,y,w,h = cv2.boundingRect(cnt)
            org = (x+5,y-5) 
            cv2.rectangle(img_original,(x,y),(x+w,y+h),(0,255,0),2)
            cv2.rectangle(img_original,(x,y),(x+120,y-40),(0,255,0),-1)
            #cv2.putText(img_original,"TUMOR",(x+10,y-10),cv2.FONT_HERSEY_SIMPLEX,1,(255,255,255),2)
            cv2.putText(img_original, 'TUMOR', org, font, fontScale, color, thickness, cv2.LINE_AA)


    cv2.imshow('LIVE',img_original)
    cv2.waitKey(0)
    #cv2.destroyAllWindow()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import os

training_dir = r"Dataset/Dataset Citra 240 - 235/New Dataset/train/"
validation_dir = r"Dataset/Dataset Citra 240 - 235/New Dataset/val/"

categories = ["glioma_tumor","meningioma_tumor","no_tumor","pituitary_tumor"]
```

```python
img_size = (150,150)

training_data = []

def create_training_data():
    for category in categories:
        path = os.path.join(training_dir,category)
        class_num = categories.index(category)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img))
                new_array = cv2.resize(img_array,img_size) 
                training_data.append([new_array,class_num])
            except Exception as e:
                pass
create_training_data()
#print("Berhasil Create Training Data")
```

```python
X_train = []
Y_train = []
for features,label in training_data:
    X_train.append(features)
    Y_train.append(label)
X_train = np.array(X_train).reshape(-1,150,150)
X_train = X_train.astype('float32')/255.0  
X_train = X_train.reshape(-1,150,150,3)
print(X_train.shape)
```

```python
IMG_SIZE = 150

validation_data = []

def create_validation_data():
    for category in categories:
        path = os.path.join(validation_dir,category)
        class_num = categories.index(category)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img))
                new_array = cv2.resize(img_array,img_size) 
                validation_data.append([new_array,class_num])
            except Exception as e:
                pass
create_validation_data()
#print("Berhasil Create Validation Data")
```

```python
X_val = []
Y_val = []
for features,label in validation_data:
    X_val.append(features)
    Y_val.append(label)
X_val = np.array(X_val).reshape(-1,150,150)
X_val = X_val.astype('float32')/255.0  
X_val = X_val.reshape(-1,150,150,3)
print(X_val.shape)
```

```python
Y_train = np.array(Y_train)
Y_val = np.array(Y_val)
```

```python
from keras.utils.np_utils import to_categorical # convert to one-hot-encoding
Y_train = to_categorical(Y_train, num_classes = 4)
Y_val = to_categorical(Y_val, num_classes = 4)
```

```python
print("x_train shape",X_train.shape)
print("x_test shape",X_val.shape)
print("y_train shape",Y_train.shape)
print("y_test shape",Y_val.shape)
```

```python
import itertools
from keras.utils.np_utils import to_categorical 
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D, Input
#from keras.optimizers import adam
import adam
from keras.preprocessing.image import ImageDataGenerator
```

```python
model = Sequential(
    [
     Input(shape=(150,150,3)),

     # Activation menggunakan relu
     Conv2D(64, (5,5),padding = 'Same', activation = 'relu'),
     MaxPooling2D(pool_size = (2,2)),
     Dropout(0.25),

     Conv2D(128, (3,3),padding = 'Same', activation = 'relu'),
     MaxPooling2D(pool_size = (2,2), strides=(2,2)),
     Dropout(0.25),

     Conv2D(128, (3,3),padding = 'Same', activation = 'relu'),
     MaxPooling2D(pool_size = (2,2), strides=(2,2)),
     Dropout(0.3),

     Conv2D(128, (3,3),padding = 'Same', activation = 'relu'),
     MaxPooling2D(pool_size = (2,2), strides=(2,2)),
     Dropout(0.3),

     Conv2D(256, (3,3),padding = 'Same', activation = 'relu'),
     MaxPooling2D(pool_size = (2,2), strides=(2,2)),
     Dropout(0.3),

      #meratakan 1 dimensi 
     Flatten(),
     
     Dense(1024, activation = 'relu'),
     Dropout(0.5),
     Dense(4, activation='softmax')
    ]
)

model.summary()
```

```python
import tensorflow as tf
opt = tf.keras.optimizers.Adam(learning_rate=0.001)
```

```python
from tensorflow.keras.optimizers import Adam 
```

```python
Adam(learning_rate=0.00146, name='Adam')
model.compile(optimizer = 'Adam',loss = 'categorical_crossentropy',metrics = ['accuracy'])
```

```python
epochs = 50
batch_size = 32

datagen = ImageDataGenerator(
        rotation_range=0,
        zoom_range = 0,
        width_shift_range=0,  
        height_shift_range=0,  
        horizontal_flip=True,  
        vertical_flip=False)  
```

```python
import tensorflow as tf
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('accuracy')>0.97 and logs.get('val_accuracy')>0.97):
      print("\n\nAkurasi train dan validasi telah mencapai nilai > 97%!")
      self.model.stop_training = True
callbacks = myCallback()
```

```python
import time

start = time.time()
datagen.fit(X_train)
history = model.fit_generator(datagen.flow(X_train,Y_train, batch_size=batch_size),
                              epochs = epochs, validation_data = (X_val,Y_val),
                              steps_per_epoch = X_train.shape[0] // batch_size,
                              callbacks=[callbacks])
stop = time.time()
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.BuPu):

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```
